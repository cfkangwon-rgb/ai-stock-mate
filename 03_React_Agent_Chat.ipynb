```python
from langchain_openai import ChatOpenAI
from langchain_teddynote.tools.tavily import TavilySearch
from langchain_core.messages import HumanMessage
from langgraph.checkpoint.memory import MemorySaver
from langgraph.prebuilt import create_react_agent

# 메모리 설정
memory = MemorySaver()

# 모델 설정
model = ChatOpenAI(model_name="gpt-4o-mini")

print("##################### Completed Cell #####################")
```

```python
from langchain_teddynote.tools.tavily import TavilySearch

web_search = TavilySearch(
    topic="general",  # 뉴스 주제 (general 또는 news)
    max_results=3,  # 최대 검색 결과
    include_answer=False,
    include_raw_content=False,
    include_images=False,
    format_output=False,  # 결과 포맷팅
)

web_search.name = "web_search"
web_search.description = "뉴스의 주제에 대해 웹을 검색하려면 이 도구를 사용하세요"

tools = [web_search]

print("##################### Completed Cell #####################")
```

```python
from langgraph.prebuilt import create_react_agent

# 1. 시스템 프롬프트 정의
system_prompt = """
    당신은 주식 투자 전문가이다.
    문의한 종목에대해 매수/매도/홀딩 의견을 주고, 관련 근거도 제시해야 한다.
    
    # 답변은 아래예시로 부탁해.
    1. 의견: 매수
    2. 근거: 
        - 근거1 (출처1)
        - 근거2 (출처2)
        - 근거3 (출처3)
    3. 종합:
"""

# 2. state_modifier 파라미터에 프롬프트 주입
agent_executor = create_react_agent(
    model, tools, checkpointer=memory, state_modifier=system_prompt
)

print("##################### Completed Cell #####################")
```

```python
from langchain_teddynote.messages import stream_graph

config = {"configurable": {"thread_id": "abc123"}}
inputs = {"messages": [("human", "SK하이닉스 주식을 매수 해도 되나요?")]}
stream_graph(agent_executor, inputs, config, node_names=["agent", "tools"])
```
